### 12.	Electric Range Correlation:
-	Is there a correlation between the electric range and the base MSRP of the vehicles? Provide the correlation coefficient and interpret the result.


```python
import pandas as pd

# Load the CSV file
file_path = 'ev.csv'  # Replace with the actual file path
ev_data = pd.read_csv(file_path)

# Convert 'Electric Range' and 'Base MSRP' to numeric, coercing errors to NaN
ev_data['Electric Range'] = pd.to_numeric(ev_data['Electric Range'], errors='coerce')
ev_data['Base MSRP'] = pd.to_numeric(ev_data['Base MSRP'], errors='coerce')

# Drop rows with missing or NaN values in 'Electric Range' or 'Base MSRP'
ev_data_clean = ev_data.dropna(subset=['Electric Range', 'Base MSRP'])

# Calculate the correlation coefficient
correlation_coefficient = ev_data_clean['Electric Range'].corr(ev_data_clean['Base MSRP'])

print(f"Correlation coefficient between Electric Range and Base MSRP: {correlation_coefficient:.2f}")

# Interpret the result
if correlation_coefficient > 0.5:
    interpretation = "There is a strong positive correlation between electric range and base MSRP."
elif correlation_coefficient > 0.3:
    interpretation = "There is a moderate positive correlation between electric range and base MSRP."
elif correlation_coefficient > 0.1:
    interpretation = "There is a weak positive correlation between electric range and base MSRP."
else:
    interpretation = "There is little to no correlation between electric range and base MSRP."

print(interpretation)
```

### 19.	Range vs. Year:
-	Is there a trend between the model year and the electric range of the vehicles? Provide a scatter plot and analyze the trend.


```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Drop rows with missing values in 'Model Year' or 'Electric Range'
df_clean = df.dropna(subset=['Model Year', 'Electric Range'])

# Convert 'Model Year' and 'Electric Range' to numeric values (if they are not already)
df_clean['Model Year'] = pd.to_numeric(df_clean['Model Year'], errors='coerce')
df_clean['Electric Range'] = pd.to_numeric(df_clean['Electric Range'], errors='coerce')

# Drop rows with NaN values after conversion
df_clean = df_clean.dropna(subset=['Model Year', 'Electric Range'])

# Plot the scatter plot of model year vs. electric range
plt.figure(figsize=(14, 8))
plt.scatter(df_clean['Model Year'], df_clean['Electric Range'], alpha=0.5, label='Data Points')

# Fit a regression line to observe the trend
slope, intercept, r_value, p_value, std_err = linregress(df_clean['Model Year'], df_clean['Electric Range'])
plt.plot(df_clean['Model Year'], intercept + slope * df_clean['Model Year'], color='red', label=f'Fit Line (slope={slope:.2f})')

# Annotate the plot
plt.title('Model Year vs. Electric Range of Vehicles')
plt.xlabel('Model Year')
plt.ylabel('Electric Range (miles)')
plt.legend()
plt.grid(True)
plt.show()

# Print the regression line details
print(f'Slope: {slope}')
print(f'Intercept: {intercept}')
print(f'R-squared: {r_value**2}')
```

### 20.	Legislative District and MSRP:
-	How does the average base MSRP vary across different legislative districts?


```python
import pandas as pd
import matplotlib.pyplot as plt

# Drop rows with missing values in 'Legislative District' or 'Base MSRP'
df_clean = df.dropna(subset=['Legislative District', 'Base MSRP'])

# Convert 'Base MSRP' to numeric values (if it is not already)
df_clean['Base MSRP'] = pd.to_numeric(df_clean['Base MSRP'], errors='coerce')

# Drop rows with NaN values after conversion
df_clean = df_clean.dropna(subset=['Base MSRP'])

# Group by legislative district and calculate the average base MSRP
avg_msrp_by_district = df_clean.groupby('Legislative District')['Base MSRP'].mean().sort_values()

# Plot the average base MSRP for each legislative district
plt.figure(figsize=(14, 8))
avg_msrp_by_district.plot(kind='bar', color='skyblue')
plt.title('Average Base MSRP by Legislative District')
plt.xlabel('Legislative District')
plt.ylabel('Average Base MSRP ($)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
